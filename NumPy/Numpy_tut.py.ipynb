#### Faster than list

```python
import datetime

my_arr = np.arange(10000)
my_list = list(my_arr)
start = datetime.datetime.now()    
my_arr2 = my_arr * 2
print(datetime.datetime.now() - start)
start = datetime.datetime.now()    
my_list2 = [x * 2 for x in my_list]
print(datetime.datetime.now() - start)
```

#### Arithmetic operations

#### Basic Indexing and Slicing

```python
arr = np.arange(10)
print(arr[5])
print(arr[:5], arr[1:5])
arr[:] = 0
print(arr)

print('='*64)

# numpy does not copy data
arr2 = arr[1:2]
arr2[:] = 1
print(arr2)
print(arr)

print('='*64)

arr2d = np.array([[1,2,3], [4,5,6]])
print(arr2d[1], arr2d[1,1])
print(arr2d[:,1])
```

#### Boolean Indexing

```python
names = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'])
cond = names == 'Bob'
print(cond)
data = np.random.randn(7, 5)
print(data)
print('='*64)
print(data[names=='Bob'])
print('='*64)
print(data[~(names=='Bob')])
print('='*64)
cond = (names == 'Bob') | (names == 'Joe')
print(cond)
print(data[cond])
print('='*64)
data[data<0] = 0
print(data)
```

#### Fancy Indexing

#### Transposing arrays and swapping axes

```python
arr = np.arange(15)
print(arr)
arr2 = arr.reshape((5,3))
print(arr2)
print(arr2.T)
print('='*64)
print(np.dot(arr2.T, arr2))
print('='*64)
print(arr2.swapaxes(0,1)) # Transposing is a special case of swapping indeces
```

#### Array Orientated Programming

```python
import matplotlib.pyplot as plt

points = np.arange(-5, 5, 0.01)
print(points)
xs, ys = np.meshgrid(points, points)
print(xs)
print(ys)
z = np.sqrt(xs**2 + ys**2)
plt.imshow(z, cmap=plt.cm.gray)
plt.colorbar()
plt.title('Image plot of $\sqrt{x^2+y^2}$ for a grid of values')
```

#### Conditional Logic

```python
xarr = np.arange(10)
print(xarr)
yarr = np.random.randn(10)
print(yarr)
cond = np.array([False]*5+[True]*5)
print(np.where(cond, xarr, yarr))

print(np.where(yarr>0, yarr, xarr))
```

#### Simple Statistical Methods

```python
arr = np.random.randn(5,4)
print(arr)
print(arr.mean())
print(arr.sum())
print('='*64)
print(arr.mean(axis=0))
print(arr.mean(axis=1))
print('='*64)
print(arr.cumsum(axis=0)) # cumulative sum across rows
print(arr.cumsum(axis=0)) # cumulative sum across columns
```

#### Linear Algebra

```python
from numpy.linalg import inv, qr

x = np.arange(10)
y = np.random.randint(10, size=(1,10))
print(x)
print(y)
print(y.dot(x))
print('='*64)
z = np.random.randint(10, size=(5,5))
print(inv(z))
print('='*64)
q, r = qr(z)
print(q)
print('='*64)
print(r)
```

#### Random Walk Simulation

```python
nsteps = 1000
draws = np.random.randint(0,2, size=nsteps)
steps = np.where(draws>0, 1, -1)
walks = steps.cumsum()
plt.plot(walks)
print((np.abs(walks)>=10).argmax())
```

```python
nwalks = 5000 # multiple random walks
nsteps = 1000
draws = np.random.randint(0,2, size=(nwalks, nsteps))
steps = np.where(draws>0, 1, -1)
walks = steps.cumsum(1)
walks
```
